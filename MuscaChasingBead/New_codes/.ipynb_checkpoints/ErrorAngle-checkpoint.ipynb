# Error angle and distance to bead data from smooth data

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Save error angle and distance data csv

```python
#Input dir- folder path of smooth csv, output dir- where the chase metric csv should be saved
INPUT_DIR  = r"./../../Tanvi-files/xyz_Smooth"
OUTPUT_DIR = r"./../../Tanvi-files/Error_Angles"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Error angle functions
def compute_error_angle_center(bead, head, center):
    v_heading = head - center
    v_bead    = head - bead
    dot = np.einsum("ij,ij->i", v_heading, v_bead)
    nh = np.linalg.norm(v_heading, axis=1)
    nb = np.linalg.norm(v_bead, axis=1)
    cos = np.clip(dot / (nh * nb), -1, 1)
    return np.degrees(np.arccos(cos))

def compute_error_angle_head(bead, head):
    v_bead = bead - head
    forward = np.gradient(head, axis=0)
    dot = np.einsum("ij,ij->i", forward, v_bead)
    nf = np.linalg.norm(forward, axis=1)
    nb = np.linalg.norm(v_bead, axis=1)
    cos = np.clip(dot / (nf * nb), -1, 1)
    return np.degrees(np.arccos(cos))

# Find smooth files
csv_files = glob.glob(os.path.join(INPUT_DIR, "*_SMOOTH.csv"))
if not csv_files:
    raise FileNotFoundError("No SMOOTH CSV files found.")


for path in csv_files:
    fname = os.path.basename(path)
    print("Processing:", fname)

    df = pd.read_csv(path)

    # Use frame directly from CSV
    frames = df["frame"].values

    bead   = df[["pt1_X","pt1_Y","pt1_Z"]].values
    head   = df[["pt2_X","pt2_Y","pt2_Z"]].values
    center = df[["center_X","center_Y","center_Z"]].values

    # Distances
    dist_center = np.linalg.norm(bead - center, axis=1)
    dist_head   = np.linalg.norm(bead - head, axis=1)

    # Error angles
    err_center = compute_error_angle_center(bead, head, center)
    err_head   = compute_error_angle_head(bead, head)

    # Save CSV
    out_df = pd.DataFrame({
        "frame": frames,
        "dist_center_m": dist_center,
        "dist_head_m": dist_head,
        "error_angle_center_deg": err_center,
        "error_angle_head_deg": err_head
    })

    out_name = fname.replace("_SMOOTH.csv", "_CHASE_METRICS.csv")
    out_df.to_csv(os.path.join(OUTPUT_DIR, out_name), index=False)

    print("Saved:", out_name)

print("\n Chase metrics CSVs saved.")
```

#save and display plots from error angle and distance to bead

```python

# Where CHASE_METRICS CSVs live
INPUT_DIR = r"D:\Nishka_Analysis\Analysis_Body_Wing\Data_DLC_DlTdv8\DLTdv8_Data\Final_DLTdv87_Data\Four_Chase_CSVData\Error_Angles"

# Base directory containing Trial2_200rpm, Trial3_180rpm, etc.
BASE_TRIAL_DIR = r"D:\Nishka_Analysis\Analysis_Body_Wing\Four_Chase_Data"

# ==========================================================
# FIND CHASE METRICS FILES
# ==========================================================

csv_files = sorted(glob.glob(os.path.join(INPUT_DIR, "*_CHASE_METRICS.csv")))
if not csv_files:
    raise FileNotFoundError("No CHASE_METRICS CSV files found.")

# ==========================================================
# PLOTTING
# ==========================================================

for path in csv_files:
    base = os.path.splitext(os.path.basename(path))[0]
    print("Plotting:", base)

    # ------------------------------------------------------
    # Extract trial name (e.g. Trial2_200rpm)
    # ------------------------------------------------------
    try:
        trial_name = [p for p in base.split("_") if p.startswith("Trial")][0]
    except IndexError:
        raise ValueError(f"Could not extract trial name from: {base}")

    # ------------------------------------------------------
    # Create trial-specific CHASE_PLOTS folder
    # ------------------------------------------------------
    trial_dir = os.path.join(BASE_TRIAL_DIR, trial_name)
    save_dir = os.path.join(trial_dir, "CHASE_PLOTS")
    os.makedirs(save_dir, exist_ok=True)

    # ------------------------------------------------------
    # Load data
    # ------------------------------------------------------
    df = pd.read_csv(path)
    t = df["frame"].values

    # ================= CENTER-BASED =================
    fig, ax1 = plt.subplots(figsize=(14, 6))

    ax1.plot(t, df["dist_center_m"], color="darkred", linewidth=2)
    ax1.set_xlabel("Frame")
    ax1.set_ylabel("Distance (m)", color="darkred")
    ax1.tick_params(axis="y", labelcolor="darkred")
    ax1.grid(True)

    ax2 = ax1.twinx()
    ax2.plot(t, df["error_angle_center_deg"], color="purple", linewidth=2)
    ax2.set_ylabel("Error Angle (°)", color="purple")
    ax2.tick_params(axis="y", labelcolor="purple")

    plt.title("CENTER-Based: Distance to Bead + Error Angle")
    plt.tight_layout()

    center_out = os.path.join(save_dir, f"{base}_CENTER.png")
    plt.savefig(center_out, dpi=300)
    plt.close()

    # ================= HEAD-BASED =================
    fig, ax1 = plt.subplots(figsize=(14, 6))

    ax1.plot(t, df["dist_head_m"], color="goldenrod", linewidth=2)
    ax1.set_xlabel("Frame")
    ax1.set_ylabel("Distance (m)", color="goldenrod")
    ax1.tick_params(axis="y", labelcolor="goldenrod")
    ax1.grid(True)

    ax2 = ax1.twinx()
    ax2.plot(t, df["error_angle_head_deg"], color="blue", linewidth=2)
    ax2.set_ylabel("Error Angle (°)", color="blue")
    ax2.tick_params(axis="y", labelcolor="blue")

    plt.title("HEAD-Based: Distance to Bead + Error Angle")
    plt.tight_layout()

    head_out = os.path.join(save_dir, f"{base}_HEAD.png")
    plt.savefig(head_out, dpi=300)
    plt.close()

    print("Saved to:", save_dir)

print("\n Chase metrics plots saved trial-wise.")
```

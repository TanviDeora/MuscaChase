# 3D plot of head-bead and center-bead from smooth data 

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm
from mpl_toolkits.mplot3d import Axes3D
```

# 3D plot without hue

```python

# Where smooth CSVs live
INPUT_DIR = r"D:\Nishka_Analysis\Analysis_Body_Wing\Data_DLC_DlTdv8\DLTdv8_Data\Final_DLTdv87_Data\Four_Chase_CSVData\xyz_Smooth"

# Base directory containing Trial2_200rpm, Trial3_180rpm, etc.
BASE_TRIAL_DIR = r"D:\Nishka_Analysis\Analysis_Body_Wing\Four_Chase_Data"

# ==========================================================
# FIND SMOOTH FILES
# ==========================================================

csv_files = sorted(glob.glob(os.path.join(INPUT_DIR, "*_SMOOTH.csv")))
if not csv_files:
    raise FileNotFoundError("No SMOOTH CSV files found.")

# ==========================================================
# PLOTTING
# ==========================================================

for path in csv_files:
    base = os.path.splitext(os.path.basename(path))[0]
    print("Plotting:", base)

    # ------------------------------------------------------
    # Extract trial name (e.g. Trial2_200rpm)
    # ------------------------------------------------------
    try:
        trial_name = [p for p in base.split("_") if p.startswith("Trial")][0]
    except IndexError:
        raise ValueError(f"Could not extract trial name from: {base}")

    # ------------------------------------------------------
    # Create trial-specific TRAJECTORY_3D_PLOTS folder
    # ------------------------------------------------------
    trial_dir = os.path.join(BASE_TRIAL_DIR, trial_name)
    save_dir = os.path.join(trial_dir, "TRAJECTORY_3D_PLOTS")
    os.makedirs(save_dir, exist_ok=True)

    # ------------------------------------------------------
    # Load data
    # ------------------------------------------------------
    df = pd.read_csv(path)

    head   = df[["pt2_X", "pt2_Y", "pt2_Z"]].values
    bead   = df[["pt1_X", "pt1_Y", "pt1_Z"]].values
    center = df[["center_X", "center_Y", "center_Z"]].values

    # ==================================================
    # PLOT 1 — HEAD + BEAD
    # ==================================================
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection="3d")

    ax.plot(
        head[:, 0], head[:, 1], head[:, 2],
        color="blue", linewidth=2, label="Head"
    )
    ax.plot(
        bead[:, 0], bead[:, 1], bead[:, 2],
        color="red", linewidth=2, label="Bead"
    )

    ax.set_title(f"{base} — 3D Trajectory: Head + Bead")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.legend()

    plt.tight_layout()
    out1 = os.path.join(save_dir, f"{base}_3D_Head_Bead.png")
    plt.savefig(out1, dpi=300)
    plt.close()

    # ==================================================
    # PLOT 2 — CENTER + BEAD
    # ==================================================
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection="3d")

    ax.plot(
        center[:, 0], center[:, 1], center[:, 2],
        color="green", linewidth=2, label="Body Center"
    )
    ax.plot(
        bead[:, 0], bead[:, 1], bead[:, 2],
        color="red", linewidth=2, label="Bead"
    )

    ax.set_title(f"{base} — 3D Trajectory: Center + Bead")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")
    ax.legend()

    plt.tight_layout()
    out2 = os.path.join(save_dir, f"{base}_3D_Center_Bead.png")
    plt.savefig(out2, dpi=300)
    plt.close()

    print("Saved to:", save_dir)

print("\n 3D trajectory plots saved trial-wise.")

```

# 3D plot with hue

```python

# Where smooth CSVs live
INPUT_DIR = r"D:\Nishka_Analysis\Analysis_Body_Wing\Data_DLC_DlTdv8\DLTdv8_Data\Final_DLTdv87_Data\Four_Chase_CSVData\xyz_Smooth"

# Base directory containing Trial2_200rpm, Trial3_180rpm, etc.
BASE_TRIAL_DIR = r"D:\Nishka_Analysis\Analysis_Body_Wing\Four_Chase_Data"

# ==========================================================
# FIND SMOOTH FILES
# ==========================================================

csv_files = sorted(glob.glob(os.path.join(INPUT_DIR, "*_SMOOTH.csv")))
if not csv_files:
    raise FileNotFoundError("No SMOOTH CSV files found.")

# ==========================================================
# PLOTTING
# ==========================================================

for path in csv_files:
    base = os.path.splitext(os.path.basename(path))[0]
    print("Plotting:", base)

    # ------------------------------------------------------
    # Extract trial name (e.g. Trial2_200rpm)
    # ------------------------------------------------------
    try:
        trial_name = [p for p in base.split("_") if p.startswith("Trial")][0]
    except IndexError:
        raise ValueError(f"Could not extract trial name from: {base}")

    # ------------------------------------------------------
    # Create trial-specific TRAJECTORY_3D_HUE_PLOTS folder
    # ------------------------------------------------------
    trial_dir = os.path.join(BASE_TRIAL_DIR, trial_name)
    save_dir = os.path.join(trial_dir, "TRAJECTORY_3D_HUE_PLOTS")
    os.makedirs(save_dir, exist_ok=True)

    # ------------------------------------------------------
    # Load data
    # ------------------------------------------------------
    df = pd.read_csv(path)

    head = df[["pt2_X", "pt2_Y", "pt2_Z"]].values
    bead = df[["pt1_X", "pt1_Y", "pt1_Z"]].values

    n_frames = head.shape[0]
    colors = np.linspace(0, 1, n_frames)

    cmap_head = cm.get_cmap("viridis")
    cmap_bead = cm.get_cmap("plasma")

    # ==================================================
    # 3D PLOT WITH HUE
    # ==================================================
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection="3d")

    # ---- Head trajectory ----
    for i in range(n_frames - 1):
        ax.plot(
            head[i:i+2, 0],
            head[i:i+2, 1],
            head[i:i+2, 2],
            color=cmap_head(colors[i]),
            linewidth=3
        )

    # ---- Bead trajectory ----
    for i in range(n_frames - 1):
        ax.plot(
            bead[i:i+2, 0],
            bead[i:i+2, 1],
            bead[i:i+2, 2],
            color=cmap_bead(colors[i]),
            linewidth=3
        )

    # --------------------------------------------------
    # LABELS
    # --------------------------------------------------
    ax.set_title(f"{base} — 3D Trajectory (Hue = Time)", fontsize=14)
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_zlabel("Z")

    # --------------------------------------------------
    # COLORBARS
    # --------------------------------------------------
    sm_head = cm.ScalarMappable(cmap=cmap_head)
    sm_head.set_array([])
    cbar1 = plt.colorbar(sm_head, ax=ax, fraction=0.03, pad=0.10)
    cbar1.set_label("Frame progression (Head)")

    sm_bead = cm.ScalarMappable(cmap=cmap_bead)
    sm_bead.set_array([])
    cbar2 = plt.colorbar(sm_bead, ax=ax, fraction=0.03, pad=0.02)
    cbar2.set_label("Frame progression (Bead)")

    plt.tight_layout()

    out_path = os.path.join(save_dir, f"{base}_3D_HUE_Head_Bead.png")
    plt.savefig(out_path, dpi=300)
    plt.close()

    print("Saved to:", save_dir)

print("\n Hue-encoded 3D trajectory plots saved trial-wise.")
```

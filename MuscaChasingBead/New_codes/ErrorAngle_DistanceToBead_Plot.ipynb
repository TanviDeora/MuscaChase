```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#save and display plots from error angle and distance to bead
```

```python
# Where CHASE_METRICS CSVs live
INPUT_DIR = r"./../../dataFolders/MuscaChasingBeads/ErrorAngle_DistFromBead/"
fig_output = r"./../../dataFolders/MuscaChasingBeads/Figures/"

# ==========================================================
# FIND CHASE METRICS FILES
# ==========================================================

csv_files = sorted(glob.glob(os.path.join(INPUT_DIR, "*_CHASE_METRICS.csv")))
if not csv_files:
    raise FileNotFoundError("No CHASE_METRICS CSV files found.")

# ==========================================================
# PLOTTING
# ==========================================================

for path in csv_files:
    base = os.path.splitext(os.path.basename(path))[0]
    print("Plotting:", base)

    # ------------------------------------------------------
    # Extract trial name (e.g. Trial2_200rpm)
    # ------------------------------------------------------
    trial_name = '_'.join(base.split("_")[0:2])[:-6]


    # ------------------------------------------------------
    # Create trial-specific CHASE_PLOTS folder
    # ------------------------------------------------------
#     trial_dir = os.path.join(FIG_OUTPUT, trial_name)
#     save_dir = os.path.join(trial_dir, "CHASE_PLOTS")
#     os.makedirs(save_dir, exist_ok=True)

    # ------------------------------------------------------
    # Load data
    # ------------------------------------------------------
    df = pd.read_csv(path)
    t = df["frame"].values

    # ================= CENTER-BASED =================
    fig, ax1 = plt.subplots(figsize=(14, 6))

    ax1.plot(t, df["dist_center_m"], color="darkred", linewidth=2)
    ax1.set_xlabel("Frame")
    ax1.set_ylabel("Distance (m)", color="darkred")
    ax1.tick_params(axis="y", labelcolor="darkred")
    ax1.grid(True)

    ax2 = ax1.twinx()
    ax2.plot(t, df["error_angle_center_deg"], color="purple", linewidth=2)
    ax2.set_ylabel("Error Angle (°)", color="purple")
    ax2.tick_params(axis="y", labelcolor="purple")

    plt.title("CENTER-Based: Distance to Bead + Error Angle")
    plt.tight_layout()

    center_out = os.path.join(fig_output, f"{trial_name}_CENTER.png")
    plt.savefig(center_out, dpi=300)
    plt.close()

    # ================= HEAD-BASED =================
    fig, ax1 = plt.subplots(figsize=(14, 6))

    ax1.plot(t, df["dist_head_m"], color="goldenrod", linewidth=2)
    ax1.set_xlabel("Frame")
    ax1.set_ylabel("Distance (m)", color="goldenrod")
    ax1.tick_params(axis="y", labelcolor="goldenrod")
    ax1.grid(True)

    ax2 = ax1.twinx()
    ax2.plot(t, df["error_angle_head_deg"], color="blue", linewidth=2)
    ax2.set_ylabel("Error Angle (°)", color="blue")
    ax2.tick_params(axis="y", labelcolor="blue")

    plt.title("HEAD-Based: Distance to Bead + Error Angle")
    plt.tight_layout()

    head_out = os.path.join(fig_output, f"{trial_name}_HEAD.png")
    plt.savefig(head_out, dpi=300)
    plt.close()

    print("Saved to:", fig_output)

print("\n Chase metrics plots saved trial-wise.")
```

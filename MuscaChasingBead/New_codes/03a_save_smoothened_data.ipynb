```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#save and display smooth plots

```python

# Where smooth CSVs live
INPUT_DIR = r"./../../Tanvi-files/xyz_Smooth"

# Base directory containing Trial2_200rpm, Trial3_180rpm, etc.
BASE_TRIAL_DIR = r"./../../Tanvi-files/Figures/"

# ==========================================================
# POINT DEFINITIONS
# ==========================================================

points = {
    "pt1": "Bead",
    "pt2": "Head",
    "pt3": "Left Wing Base",
    "pt4": "Right Wing Base",
    "pt5": "Left Wing Tip",
    "pt6": "Right Wing Tip",
    "pt7": "Abdomen Tip",
    "center": "Center"
}

# ==========================================================
# FIND SMOOTH FILES
# ==========================================================

csv_files = sorted(glob.glob(os.path.join(INPUT_DIR, "*_SMOOTH.csv")))
if not csv_files:
    raise FileNotFoundError("No SMOOTH CSV files found.")

# ==========================================================
# PLOTTING
# ==========================================================

for path in csv_files:
    base = os.path.splitext(os.path.basename(path))[0]
    print("Plotting:", base)

    # ------------------------------------------------------
    # Extract trial name (e.g. Trial2_200rpm)
    # ------------------------------------------------------
    trial_name = '_'.join(base.split("_")[0:2])[:-6]
#     try:
#         trial_name = [p for p in base.split("_") if p.startswith("Trial")][0]
#     except IndexError:
#         raise ValueError(f"Could not extract trial name from: {base}")

    # ------------------------------------------------------
    # Create trial-specific SMOOTH_PLOTS folder
    # ------------------------------------------------------
    trial_dir = os.path.join(BASE_TRIAL_DIR, trial_name)
    save_dir = os.path.join(trial_dir, "SMOOTH_PLOTS")
    os.makedirs(save_dir, exist_ok=True)

    # ------------------------------------------------------
    # Load data
    # ------------------------------------------------------
    df = pd.read_csv(path)
    frames = df["frame"].values

    # ------------------------------------------------------
    # Collect smoothed XYZ data
    # ------------------------------------------------------
    smooth = {}
    for pt in points.keys():
        smooth[pt] = np.column_stack([
            df[f"{pt}_X"].values,
            df[f"{pt}_Y"].values,
            df[f"{pt}_Z"].values,
        ])

    # ------------------------------------------------------
    # Plot XYZ vs frame
    # ------------------------------------------------------
    for pt, label in points.items():
        for i, axis in enumerate(["X", "Y", "Z"]):

            plt.figure(figsize=(10, 4))
            plt.plot(frames, smooth[pt][:, i], linewidth=2, color="green")
            plt.title(f"{label} ({pt}) — SMOOTHED — {axis}")
            plt.xlabel("Frame")
            plt.ylabel(axis)
            plt.grid(True)
            plt.tight_layout()

            out_name = f"{base}_{pt}_{axis}_SMOOTH.png"
            plt.savefig(os.path.join(save_dir, out_name), dpi=300)
            plt.close()

    print(f"Saved plots to: {save_dir}")

print(" All smoothed plots (including center) saved trial-wise.")
```

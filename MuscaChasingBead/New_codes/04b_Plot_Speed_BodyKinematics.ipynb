```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## save and display speed plots

```python
# create folder to keep figures
# figure output directory
fig_dir = r"./../../dataFolders/MuscaChasingBeads/Figures/Speed/"
os.makedirs(fig_dir, exist_ok=True)
```

```python
# Where SPEED CSVs live
INPUT_DIR = r"./../../dataFolders/MuscaChasingBeads/Speed/"


# FIND SPEED FILES
csv_files = sorted(glob.glob(os.path.join(INPUT_DIR, "*_SPEED.csv")))
if not csv_files:
    raise FileNotFoundError("No SPEED CSV files found.")

# Plot
for path in csv_files:
    base = os.path.splitext(os.path.basename(path))[0]
    print("Plotting:", base)

    # ------------------------------------------------------
    # Extract trial name 
    # ------------------------------------------------------
    trial_name = '_'.join(base.split("_")[0:2])[:-6]

    # ------------------------------------------------------
    # Load data
    # ------------------------------------------------------
    df = pd.read_csv(path)

    # -------- FRAME HANDLING --------
    if 'frame' in df.columns:
        frames = df['frame'].values
    else:
        df = df.reset_index(drop=True)
        df['frame'] = np.arange(len(df))
        frames = df['frame'].values
    # -------------------------------------

    speed = df["speed"].values

    # ------------------------------------------------------
    # Plot
    # ------------------------------------------------------
    plt.figure(figsize=(12, 4))
    plt.plot(frames, speed, linewidth=2)

    plt.xlabel("Frame Number", fontsize=24)
    plt.ylabel("Speed (m/s)", fontsize=24)
    plt.tick_params(axis="both", labelsize=18)
    plt.grid(True)
    plt.tight_layout()

    out_path = os.path.join(fig_dir, f"{trial_name}_speed.png")
    plt.savefig(out_path, dpi=300)
    plt.close()

    print("Saved:", out_path)

print("\nSpeed plots saved trial-wise.")
```

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ── PATHS ─────────────────────────────────────────────────────────────
INPUT_DIR = r"./../../dataFolders/MuscaChasingBeads/Speed/Speed_cleaned"
FIG_DIR   = r"./../../dataFolders/MuscaChasingBeads/Figures/Speed/Speed_cleaned"
os.makedirs(FIG_DIR, exist_ok=True)

# ── FIND FILES ────────────────────────────────────────────────────────
csv_files = sorted(glob.glob(os.path.join(INPUT_DIR, "*_SPEED_CLEAN.csv")))
if not csv_files:
    raise FileNotFoundError("No CLEAN SPEED CSV files found.")

# ── LOOP ─────────────────────────────────────────────────────────────
for path in csv_files:
    base = os.path.splitext(os.path.basename(path))[0]
    print("Plotting:", base)

    trial_name = '_'.join(base.split("_")[0:2])

    df = pd.read_csv(path)

    # ---- FRAME ----
    if 'frame' in df.columns:
        frames = df['frame'].values
    else:
        df = df.reset_index(drop=True)
        frames = np.arange(len(df))

    # ---- SPEED ----
    speed = df["speed"].values

    # ------------------------------------------------------
    # Plot (no masking)
    # ------------------------------------------------------
    plt.figure(figsize=(12, 4))

    plt.plot(frames, speed, linewidth=2)

    plt.xlabel("Frame Number", fontsize=24)
    plt.ylabel("Speed", fontsize=24)
    plt.tick_params(axis="both", labelsize=18)
    plt.grid(True)

    plt.tight_layout()

    out_path = os.path.join(FIG_DIR, f"{trial_name}_speed_clean.png")
    plt.savefig(out_path, dpi=300)
    plt.close()

    print("Saved:", out_path)

print("\nClean speed plots saved.")
```

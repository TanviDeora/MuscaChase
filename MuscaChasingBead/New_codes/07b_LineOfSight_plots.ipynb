```python
import os
import re
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# where the line of sight files are located
INPUT_DIR = r"./../../dataFolders/MuscaChasingBeads/line_of_sight"
FIG_BASE_DIR = r"./../../dataFolders/MuscaChasingBeads/Figures/Line_of_sight/"

# Font size settings for readability
TITLE_SIZE = 24
LABEL_SIZE = 20
TICK_SIZE = 16

csv_files = sorted(glob.glob(os.path.join(INPUT_DIR, "*.csv")))

# ==========================================================
# PLOTTING
# ==========================================================
for path in csv_files:
    fname = os.path.basename(path)
    
    # Extract Trial Name
    match = re.search(r"(Trial\d+_\d+rpm)", fname)
    trial_name = match.group(1) if match else "Unknown_Trial"
    
    # Define the directory: FIG_BASE_DIR/Trial_X
    # (No extra "los" folder inside)
    trial_save_dir = os.path.join(FIG_BASE_DIR, trial_name)
    if not os.path.exists(trial_save_dir):
        os.makedirs(trial_save_dir)

    print(f"Generating 4 LOS plots for: {trial_name}")

    df = pd.read_csv(path)
    frames = df["frame"].values
    dist = df["distance_to_bead_m"].values

    # Helper function for large-font separate plots
    def save_styled_los(x_data, y_data, title, xlabel, ylabel, suffix, invert_x=False):
        plt.figure(figsize=(14, 8)) # Slightly taller to accommodate large title/labels
        plt.plot(x_data, y_data, color="darkblue", linewidth=2.5)
        
        plt.title(f"{title}\n{trial_name}", fontsize=TITLE_SIZE, fontweight='bold', pad=20)
        plt.xlabel(xlabel, fontsize=LABEL_SIZE)
        plt.ylabel(ylabel, fontsize=LABEL_SIZE)
        
        plt.xticks(fontsize=TICK_SIZE)
        plt.yticks(fontsize=TICK_SIZE)
        plt.grid(True, linestyle='--', alpha=0.6)
        
        if invert_x:
            plt.gca().invert_xaxis()
            
        # bbox_inches='tight' is the secret sauce for fixing cut-off labels
        plt.savefig(os.path.join(trial_save_dir, f"{trial_name}_{suffix}.png"), 
                    dpi=300, 
                    bbox_inches='tight')
        plt.close()

    # Plot 1: Angular Error vs Frame
    save_styled_los(frames, df["angular_error_deg"], "Angular Error", "Frame", "Error (deg)", "1_Angular_Error")

    # Plot 2: Cone Angle vs Frame
    save_styled_los(frames, df["cone_angle_deg"], "Cone Opening Angle", "Frame", "Angle (deg)", "2_Cone_Angle_Frame")

    # Plot 3: Cone Angle vs Distance (Inverted X)
    save_styled_los(dist, df["cone_angle_deg"], "Cone Angle vs Distance", "Distance (m)", "Angle (deg)", "3_Cone_Angle_Dist", invert_x=True)

    # Plot 4: Solid Angle vs Frame
    save_styled_los(frames, df["solid_angle_deg2"], "3D Visual Solid Angle", "Frame", "Solid Angle (deg²)", "4_Solid_Angle")

print(f"\nAll LOS plots saved directly in trial folders under: {FIG_BASE_DIR}")
```

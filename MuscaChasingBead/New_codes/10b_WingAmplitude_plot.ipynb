```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Input (wing kinematics files)
INPUT_DIR = r"./../../dataFolders/MuscaChasingBeads/Wing_Dynamics"

# Output (figures)
fig_dir = r"./../../dataFolders/MuscaChasingBeads/Figures/WingAmplitude/"
os.makedirs(fig_dir, exist_ok=True)

# Get all files
csv_files = sorted(glob.glob(os.path.join(INPUT_DIR, "*_wing_kinematics.csv")))

for path in csv_files:
    fname = os.path.basename(path)
    trial_name = fname.replace("_wing_kinematics.csv", "")

    print(f"Plotting: {trial_name}")

    df = pd.read_csv(path)

    # --- compute amplitude difference ---
    df["amp_diff"] = df["amplitude_L"] - df["amplitude_R"]

    # stroke plots
    plt.figure(figsize=(10, 4))

    plt.plot(df["frame"], df["stroke_L"], alpha=0.5, label="Stroke L")
    plt.plot(df["frame"], df["stroke_R"], alpha=0.5, label="Stroke R")

    plt.xlabel("Frame")
    plt.ylabel("Stroke Angle (deg)")
    plt.title(f"Stroke: {trial_name}")
    plt.legend()

    plt.savefig(os.path.join(fig_dir, f"{trial_name}_stroke.png"))
    plt.close()

    # amplitude plots
    plt.figure(figsize=(10, 4))

    plt.plot(df["frame"], df["amplitude_L"], linewidth=2, label="Amplitude L")
    plt.plot(df["frame"], df["amplitude_R"], linewidth=2, label="Amplitude R")

    plt.xlabel("Frame")
    plt.ylabel("Amplitude (deg)")
    plt.title(f"Amplitude: {trial_name}")
    plt.legend()

    plt.savefig(os.path.join(fig_dir, f"{trial_name}_amplitude.png"))
    plt.close()


    # 3. amplitude difference
    plt.figure(figsize=(10, 4))

    plt.plot(df["frame"], df["amp_diff"], linestyle="--")

    plt.axhline(0, linestyle=":")  # reference line

    plt.xlabel("Frame")
    plt.ylabel("Amplitude Difference (L - R)")
    plt.title(f"Amplitude Difference: {trial_name}")

    plt.savefig(os.path.join(fig_dir, f"{trial_name}_amp_diff.png"))
    plt.close()

print("\nAll plots saved successfully!")
```

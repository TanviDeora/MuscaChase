```python
import os
import glob
import pandas as pd
import numpy as np

from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
raw_tracks_path = './../../dataFolders/MuscaChasingBeads/xyz_Raw_Trimmed/'
raw_tracks = glob.glob(raw_tracks_path + '*.csv')
interpol_tracks_path = './../../dataFolders/MuscaChasingBeads/xyz_Interpolated/'
interpol_tracks = glob.glob(interpol_tracks_path + '*.csv')
smoothen_tracks_path = './../../dataFolders/MuscaChasingBeads/xyz_Smooth/'
smoothen_tracks = glob.glob(smoothen_tracks_path + '*.csv')
```

```python
# create folder to keep figures
# figure output directory
output_dir = r"./../../dataFolders/MuscaChasingBeads/Figures/Check_Smoothening/"
os.makedirs(output_dir, exist_ok=True)
```

```python
# create subplots
fig = make_subplots(
    rows=3,
    cols=1,
    shared_xaxes=True,
    vertical_spacing=0.05,
    subplot_titles=('raw', 'interp', 'smoothened')
)
```

```python
fps = 2500

for raw in raw_tracks:
    fig.data = [] 
    raw_df = pd.read_csv(raw)
    
    trialID = os.path.basename(raw)[:13]
    print(f'working on {trialID}')
    
    interpol_df = pd.read_csv([file for file in interpol_tracks if trialID in file][0])
    smoothen_df = pd.read_csv([file for file in smoothen_tracks if trialID in file][0])
    
    cols = ["center_X", "center_Y", "center_Z"]
    # assign a fixed color to each column
    colors = {"center_X": "red", "center_Y": "green", "center_Z": "blue"}
    # add traces
    for row, df in enumerate([raw_df, interpol_df, smoothen_df], start=1):   # each df → one row
        center = df.loc[:, ['center_X', 'center_Y', 'center_Z']]
        time = np.arange(0, len(center))*1/fps
        for col in cols:                        # x, y, z traces
            fig.add_trace(
                go.Scatter(
                    x=time,          # or df["time"]
                    y=df[col],
                    mode="lines",
                    name=f"{col}",
                    line=dict(color=colors[col]),
                    showlegend=(row == 1)   # legend only once      
                ),
                row=row,
                col=1
            )

    fig.update_xaxes(title_text="Time", row=3, col=1)
    fig.update_layout(height=800)     
    
    print(len(fig.data))
    fig.show()
    fig.write_html(output_dir + trialID + '.html')
    print(f'plotting done for {trialID}')
```

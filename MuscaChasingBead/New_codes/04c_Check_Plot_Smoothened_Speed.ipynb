```python
import os
import glob
import pandas as pd
import numpy as np
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
smoothen_tracks_path = r'./../../dataFolders/MuscaChasingBeads/xyz_Smooth/'
smoothen_tracks = glob.glob(smoothen_tracks_path + '*.csv')

dist_speed_path = r"./../../dataFolders/MuscaChasingBeads/Speed/"
dist_speed_files = glob.glob(dist_speed_path + '*.csv')
```

```python
# create subplots
fig = make_subplots(
    rows=4,
    cols=1,
    shared_xaxes=True,
    vertical_spacing=0.05
)
```

```python

fps = 2500

for raw in smoothen_tracks:
    fig.data = []

    # -----------------------------
    # Load data
    # -----------------------------
    smooth_df = pd.read_csv(raw)
    
    trialID = os.path.basename(raw)[:13]
    print(f'working on {trialID}')
    
    dist_speed_df = pd.read_csv(
        [file for file in dist_speed_files if trialID in file][0]
    )

    # -----------------------------
    # Time vectors (FIXED)
    # -----------------------------
    time_smooth = (
        smooth_df['frame'] / fps
        if 'frame' in smooth_df.columns
        else np.arange(len(smooth_df)) / fps
    )

    time_speed = (
        dist_speed_df['frame'] / fps
        if 'frame' in dist_speed_df.columns
        else np.arange(len(dist_speed_df)) / fps
    )

    # -----------------------------
    # Columns + colors
    # -----------------------------
    cols = ["center_X", "center_Y", "center_Z"]
    colors = {"0": "red", "1": "green", "2": "blue"}

    # -----------------------------
    # Plot raw/smoothed positions
    # -----------------------------
    for ii, col in enumerate(cols):
        fig.add_trace(
            go.Scatter(
                x=time_smooth,
                y=smooth_df[col],
                mode="lines",
                name=f"{col}",
                line=dict(color=colors[str(ii)])
            ),
            row=1,
            col=1
        )

    # -----------------------------
    # Plot processed positions (from speed CSV)
    # -----------------------------
    for ii, col in enumerate(['posx', 'posy', 'posz']):
        fig.add_trace(
            go.Scatter(
                x=time_speed,
                y=dist_speed_df[col],
                mode="lines",
                name=f"{col}",
                line=dict(color=colors[str(ii)])
            ),
            row=2,
            col=1
        )

    # -----------------------------
    # Velocity components
    # -----------------------------
    for ii, col in enumerate(['vx', 'vy', 'vz']):
        fig.add_trace(
            go.Scatter(
                x=time_speed,
                y=dist_speed_df[col],
                mode="lines",
                name=f"{col}",
                line=dict(color=colors[str(ii)])
            ),
            row=3,
            col=1
        )

    # -----------------------------
    # Speed (magnitude)
    # -----------------------------
    fig.add_trace(
        go.Scatter(
            x=time_speed,
            y=dist_speed_df['speed'],
            mode="lines",
            name="speed",
            line=dict(color='purple')
        ),
        row=4,
        col=1
    )

    # -----------------------------
    # Labels + layout
    # -----------------------------
    fig.update_yaxes(title_text='Position (m)', row=1, col=1)
    fig.update_yaxes(title_text='Position (m)', row=2, col=1)
    fig.update_yaxes(title_text='Velocity (m/s)', row=3, col=1)
    fig.update_yaxes(title_text='Speed (m/s)', row=4, col=1)

    fig.update_xaxes(title_text="Time (seconds)", row=4, col=1)
    fig.update_layout(height=800)

    # -----------------------------
    # Save
    # -----------------------------
    out_path = f"./../../dataFolders/MuscaChasingBeads/Figures/Speed/{trialID}_RawPositionAndSpeed.html"
    fig.write_html(out_path)

    print(f'plotting done for {trialID}')

print("\nAll plots generated.")
```

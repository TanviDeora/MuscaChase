```python
import os
import glob
import pandas as pd
import numpy as np

from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
smoothen_tracks_path = r'./../../Tanvi-files/xyz_Smooth/'
smoothen_tracks = glob.glob(smoothen_tracks_path + '*.csv')

dist_speed_path = r"./../../Tanvi-files/Speed/"
dist_speed_files = glob.glob(dist_speed_path + '*.csv')
```

```python
# create subplots
fig = make_subplots(
    rows=4,
    cols=1,
    shared_xaxes=True,
    vertical_spacing=0.05
)
```

```python
fps = 2500

for raw in smoothen_tracks:
    fig.data = [] 
    raw_df = pd.read_csv(raw)
    
    trialID = os.path.basename(raw)[:13]
    print(f'working on {trialID}')
    
    dist_speed_df = pd.read_csv([file for file in dist_speed_files if trialID in file][0])
    
    cols = ["center_X", "center_Y", "center_Z"]
    # assign a fixed color to each column
    colors = {"0": "red", "1": "green", "2": "blue"}
    
    # ---------------------------------------------------------------------
    # add traces
    # ---------------------------------------------------------------------
    # add x,y,z tracks in upper plot
#     center = raw_df.loc[:, ['center_X', 'center_Y', 'center_Z']]
    time = np.arange(0, len(raw_df))*1/fps
    
    # add x, y and z position
    for ii, col in enumerate(cols):                        # x, y, z traces
        fig.add_trace(
            go.Scatter(
                x=time,          # or df["time"]
                y=raw_df[col],
                mode="lines",
                name=f"{col}",
                line=dict(color=colors[str(ii)])
            ),
            row=1,
            col=1
        )
        
    # add interpolated & smoothened x, y and z position
    for ii, col in enumerate(['posx', 'posy', 'posz']):                        # x, y, z traces
        fig.add_trace(
            go.Scatter(
                x=time,          # or df["time"]
                y=dist_speed_df[col],
                mode="lines",
                name=f"{col}",
                line=dict(color=colors[str(ii)])
            ),
            row=2,
            col=1
        )
        
    # add velocity components
    for ii, col in enumerate(['vx', 'vy', 'vz']):  # x, y, z traces
        fig.add_trace(
            go.Scatter(
                x=time,          # or df["time"]
                y=dist_speed_df[col],
                mode="lines",
                name=f"{col}",
                line=dict(color=colors[str(ii)])
            ),
            row=3,
            col=1
        )    
    
        
        # add displacement and speed in bottom plot
    fig.add_trace(
        go.Scatter(
            x=time,          # or df["time"]
            y=dist_speed_df['speed'],
            mode="lines",
            name="speed",
            line=dict(color='purple')
        ),
        row=4,
        col=1
    )
      
    fig.update_yaxes(title_text='Position (m)', row=1, col=1)
    fig.update_yaxes(title_text='Position (m)', row=2, col=1)
    fig.update_yaxes(title_text='Velocity (m/s)', row=3, col=1)
    fig.update_yaxes(title_text='Velocity Magnitude (m/s²)', row=4, col=1)
    fig.update_xaxes(title_text="Time (second)", row=3, col=1)
    fig.update_layout(height=800)     
    fig.write_html("./../../Tanvi-files/Figures/Check_Smoothening/" + trialID + 'distance_speed.html')
    print(f'plotting done for {trialID}')
```

# get fft for positions and see if we see ocillations there

```python
def compute_fft(x, fs):
    x = x - np.mean(x)          # remove DC
    fft_vals = np.fft.rfft(x)
    freqs = np.fft.rfftfreq(len(x), d=1/fs)
    amp = np.abs(fft_vals)
    return freqs, amp
```

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#save and display speed plots
```

```python
# Where SPEED CSVs live
INPUT_DIR = r"./../../Tanvi-files/Speed"

# Base directory containing Trial2_200rpm, Trial3_180rpm, etc.
FIG_OUTPUT = r"./../../Tanvi-files/Figures/"

# FIND SPEED FILES

csv_files = sorted(glob.glob(os.path.join(INPUT_DIR, "*_SPEED.csv")))
if not csv_files:
    raise FileNotFoundError("No SPEED CSV files found.")

# PLot

for path in csv_files:
    base = os.path.splitext(os.path.basename(path))[0]
    print("Plotting:", base)

    # ------------------------------------------------------
    # Extract trial name (e.g. Trial2_200rpm)
    # ------------------------------------------------------
    trial_name = '_'.join(base.split("_")[0:2])[:-6]
#     try:
#         trial_name = [p for p in base.split("_") if p.startswith("Trial")][0]
#     except IndexError:
#         raise ValueError(f"Could not extract trial name from: {base}")

    # ------------------------------------------------------
    # Create trial-specific SPEED_PLOTS folder
    # ------------------------------------------------------
    trial_dir = os.path.join(FIG_OUTPUT, trial_name)
    save_dir = os.path.join(trial_dir, "SPEED_PLOTS")
    os.makedirs(save_dir, exist_ok=True)

    # ------------------------------------------------------
    # Load data
    # ------------------------------------------------------
    df = pd.read_csv(path)
    frames = df["frame"].values
    speed  = df["speed_m_per_s"].values

    # ------------------------------------------------------
    # Plot
    # ------------------------------------------------------
    plt.figure(figsize=(12, 4))
    plt.plot(frames, speed, linewidth=2)

    plt.xlabel("Frame", fontsize=24)
    plt.ylabel("Speed (m/s)", fontsize=24)
    plt.tick_params(axis="both", labelsize=18)
    plt.grid(True)
    plt.tight_layout()

    out_path = os.path.join(save_dir, f"{base}.png")
    plt.savefig(out_path, dpi=300)
    plt.close()

    print("Saved:", out_path)

print("\n Speed plots saved trial-wise.")
```

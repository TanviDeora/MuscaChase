```python
import pandas as pd
import numpy as np
import glob

import plotly.graph_objects as go
# import plotly.express as px
# import matplotlib.pyplot as plt
```

```python
def trajectoryFig_curvature(fig, df, threshold, min_r, max_r):
    
    chasee_colors = df['chasee_Trajec_curv'].apply(lambda x: 1 if x<=threshold else 2)
    chaser_colors = df['chaser_Trajec_curv'].apply(lambda x: 1 if x<=threshold else 2)
    
    fig.add_trace(go.Scatter3d(x=df['chasee_com_x'], y=df['chasee_com_y'], z=df['chasee_com_z'],
                               mode = 'markers',
                               marker = dict(
                                   size = 4,
                                   color=chasee_colors
#                                    , colorscale='YlOrRd'
#                                    , cmin = min_r, cmax = max_r
                               )
                                              ))

    fig.add_trace(go.Scatter3d(x=df['chaser_com_x'], y=df['chaser_com_y'], z=df['chaser_com_z'],
                mode = 'markers'
                , marker = dict(
                    size = 4
                    , color=chaser_colors
#                     , colorscale='YlOrRd'
#                     , colorbar=dict(thickness=20)
#                     , cmin = min_r, cmax = max_r
                )
                              ))
```

```python
files = glob.glob('./Data/Outputs/curvature/*len_*_curvature.csv')
```

```python
files
```

```python
new_list = [f for f in files if 'len_13' in f]
new_list
```

```python
c = []
for f in new_list:
    df = pd.read_csv(f, index_col = 0)
    c.extend(df.loc[:,['chasee_Trajec_curv','chaser_Trajec_curv']].values)
c = np.concatenate(c)
min_c, max_c = np.nanmin(c), np.nanmax(c)

fig = go.Figure()
fig.add_trace(go.Histogram(histfunc="count",
                           y=c))
```

```python
threshold = 0.01

for f in new_list:
    df = pd.read_csv(f, index_col = 0)
    df = df[:-13]
    name = '_'.join(f.split('\\')[-1].split('_')[:-1]) 
        
    fig = go.Figure()
    trajectoryFig_curvature(fig, df, threshold, min_c, max_c)
    fig.write_html('./Data/Figures/curvature/' + name + '_COM_trajectory_thresh.html')
```
